```python
class Nos:
    def __init__(self, valor):
        self.valor = valor
        self.direita = None
        self.esquerda = None
        self.grau = 0
        self.nivel = 0
        self.altura = 0
        self.profundidade = 0
```

```python
class Arvore:

  def __init__(self):
        self.raiz = None

  def getRaiz(self):
        return self.raiz

  def adicionar(self, valor):
    novoNo = Nos(valor)

    if(self.raiz == None):
      self.raiz = novoNo

    else:
      atual = self.raiz

      while True:
        if(novoNo.valor < atual.valor):
          if(atual.esquerda != None):
            if (atual.direita != None and atual.esquerda.esquerda != None):
              atual.altura += 1
            novoNo.nivel += 1
            novoNo.profundidade += 1
            atual = atual.esquerda
          else:
            if (atual.direita == None):
              atual.altura += 1
            atual.grau += 1
            novoNo.nivel += 1
            novoNo.profundidade += 1
            atual.esquerda = novoNo
            break
        else:
          if(atual.direita != None):
            if (atual.esquerda != None and atual.direita.direita != None):
              atual.altura += 1
            novoNo.nivel += 1
            novoNo.profundidade += 1
            atual = atual.direita
          else:
            if (atual.esquerda == None):
              atual.altura += 1
            atual.grau += 1
            novoNo.nivel += 1
            novoNo.profundidade += 1
            atual.direita = novoNo
            break

  def emOrdem(self, atual):
    if (atual != None):
      self.emOrdem(atual.esquerda)
      print(atual.valor)
      self.emOrdem(atual.direita)
    

  
  def preOrdem(self, atual):
    if (atual != None):
      print(atual.valor)
      self.preOrdem(atual.esquerda)
      self.preOrdem(atual.direita)
    

  
  def posOrdem(self, atual):
    if (atual != None):
      self.posOrdem(atual.esquerda)
      self.posOrdem(atual.direita)
      print(atual.valor)
    

  def folha(self, atual):
    if (atual != None):
      if ((atual.esquerda == None) and (atual.direita == None)):
        print(atual.valor)
      self.folha(atual.esquerda)
      self.folha(atual.direita)
    

  def grau(self, atual):
    if (atual != None):
      self.grau(atual.esquerda)
      print(f'{atual.valor} Grau: {atual.grau}')
      self.grau(atual.direita)


  def altura(self, atual):
    if (atual != None):
      self.altura(atual.esquerda)
      print(f'{atual.valor} Altura: {atual.altura}')
      self.altura(atual.direita)

  
  def profundidade(self, atual):
    if (atual != None):
      self.profundidade(atual.esquerda)
      print(f'{atual.valor} Profundidade: {atual.profundidade}')
      self.profundidade(atual.direita)
  

  def nivel(self, atual):
    if (atual != None):
      self.nivel(atual.esquerda)
      print(f'{atual.valor} Nível: {atual.nivel}')
      self.nivel(atual.direita)


  def subarvore(self, atual):
    if (atual != None):
      self.subarvore(atual.esquerda)
      if ((atual.esquerda != None) and (atual.direita != None)):
        print(f'Subárvore {atual.valor}: ({atual.valor}, {atual.esquerda.valor}, {atual.direita.valor})')
      self.subarvore(atual.direita)
```

```python
arvore = Arvore()

arvore.adicionar(10)
arvore.adicionar(5)
arvore.adicionar(13)
arvore.adicionar(11)
arvore.adicionar(15)
arvore.adicionar(9)
arvore.adicionar(4)


arvore.emOrdem(arvore.getRaiz())
print()
arvore.preOrdem(arvore.getRaiz())
print()
arvore.posOrdem(arvore.getRaiz())
print()
arvore.folha(arvore.getRaiz())
print()
arvore.grau(arvore.getRaiz())
print()
arvore.altura(arvore.getRaiz())
print()
arvore.profundidade(arvore.getRaiz())
print()
arvore.nivel(arvore.getRaiz())
print()
arvore.subarvore(arvore.getRaiz())
```
